## Q1: Graph Representation and Traversal

You are given an undirected graph with `n` nodes and `m` edges. Write code to:
- Represent it using an adjacency list.
- Perform BFS and DFS traversal starting from node 0.

### Input Format:
```
n m
u1 v1
u2 v2
...
um vm
```

### Output Format:
```
BFS: node1 node2 ...
DFS: node1 node2 ...
```


````python
from collections import deque

n, m = map(int, input().split())
adj = [[] for _ in range(n)]
for _ in range(m):
   u, v = map(int, input().split())
   adj[u].append(v)
   adj[v].append(u)

def bfs(start):
  visited = [False]*n
  q = deque()

  visited[start] = True
  q.append(start)

  result = []

  while q:
    node = q.popleft()
    result.append(node)

    for neighbor in adj[node]:
      if not visited[neighbor]:
        visited[neighbor] = True
        q.append(neighbor)

  return result

def dfs(node, visited, result):
    visited[node] = True
    result.append(node)

    for neighbor in adj[node]:
      if not visited[neighbor]:
        dfs(neighbor, visited, result)


visited = [False]*n
dfs_result = []
dfs(0, visited, dfs_result)

print("BFS", *bfs(0))
print("DFS", *dfs_result)
````


## Q2: Cycle Detection in Undirected Graph

Given an undirected graph, check whether it contains a cycle using DFS.

### Input:
Number of nodes and edges followed by edge list.

### Output:
`Yes` if cycle exists, otherwise `No`.


````python
def dfs(node, parent, adj, visited):
  visited[node] = True

  for nei in adj[node]:

    if not visited[nei]:
      if dfs(nei, node, adj, visited):
        return True

    elif nei != parent:
      return True


  return False

n, m = map(int, input().split())
adj = [[] for _ in range(n)]

for _ in range(m):
  u, v = map(int, input().split())
  adj[u].append(v)
  adj[v].append(u)

  visited = [False]*n

  cycle = False

for i in range(n):
    if not visited[i]:
      if dfs(i, -1, adj, visited)
        cycle = True
        break

if cycle:
  print("Yes")
else:
  print("No")
````


## Q3: Topological Sort (Kahn's Algorithm)

Given a **directed acyclic graph**, return any valid **topological ordering** using Kahn’s Algorithm.

### Input Format:
```
n m
u1 v1
u2 v2
...
um vm
```

### Output:
```
Topological Order: node1 node2 ...
```


````python
from collections import deque

def kts(graph):
  ind = {u: 0 for u in graph}

  for u in graph:
    for v in graph[u]:
      ind[v] += 1

  q = deque([u for u in graph if ind[u] == 0])

  order = []

  while q:
    u = q.popleft()
    order.append(u)

    for v in graph[u]:
      ind[v] -= 1
      if ind[v] == 0:
        q.append(v)

  if len(order) != len(graph):
    return "Cycle detected"

  return order

  a, b = map(int, input().split())
  graph = {i: [] for i in range(a)}

  for _ in range(b):
    u, v = map(int, input().split())
    graph[u].append(v)

  print(*kts(graph))
````


## Q4: Dijkstra’s Algorithm

Implement Dijkstra’s algorithm to find the **shortest distance from a source node (0)** to all other nodes in a graph with **non-negative edge weights**.

### Input:
```
n m
u1 v1 w1
u2 v2 w2
...
um vm wm
```

### Output:
```
0: 0
1: dist1
2: dist2
...
```


````python
import heapq

def dijkastra(n, edges):
  adj = [[] for _ in range(n)]
  for u, v, w in edges:
    adj[u].append((v, w))
    adj[v].append((u, w))

  dist = [float('inf')]*n
  dist[0] = [0]
  pq = [(0,0)]

  while pq:
    curr_dist, node = heapq.heappop(pq)

    if curr_dist > dist[node]:
      continue

    for neighbor, weight in adj[node]:
      new_dist = curr_dist + weight
      if new_dist < dist[neighbor]:
        dist[neighbor] = new_dist
        heapq.heappush(pq, (new_dist, neighbor))

  return dist





n, m = map(int, input().split())

c = []
for _ in range(m):
  c.append(list(map(int, input().split())))

print(*dijkastra(n, c))
````


## Q5: Strongly Connected Components (Kosaraju's Algorithm)

You are given a **directed graph**. Find and print all **Strongly Connected Components** (SCCs).

### Input:
```
n m
u1 v1
u2 v2
...
um vm
```

### Output:
Print each SCC in a new line.


````python
def dfs1(node, adj, visited, stack):
  visited[node] = True

  for nei in adj[node]:
    if not visited[nei]:
      dfs1(nei, adj, visited, stack)

  stack.append(node)

def dfs2(node, rev_adj, visited, component):
  visited[node] = True
  component.append(node)

  for nei in rev_adj[node]:
    if not visited[nei]:
      dfs2(nei, rev_adj, visited, component)


n, m = map(int, input().split())

adj = [[] for _ in range(n)]

rev_adj = [[] for _ in range(n)]

for _ in range(m):
  u, v = map(int, input().split())
  adj[u].append(v)
  rev_adj[v].append(u)

visited = [False]*n
stack = []

for i in range(n):
  if not visited[i]:
    dfs1(i, adj, visited, stack)

visited = [False]*n



while stack:
  node = stack.pop()

  if not visited[node]:
    component = []

    dfs2(node, rev_adj, visited, component)

    print(*component)
````
